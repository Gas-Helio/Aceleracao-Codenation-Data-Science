# Treinar e Validar

## Split

```python
from sklearn.model_selection import train_test_split
```

```python
columns_X = df_train_dummy.columns.to_list()
columns_X.remove('NU_NOTA_MT')
X = df_train_dummy[columns_X]
y = df_train_dummy[['NU_NOTA_MT']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Treino

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, ExtraTreesRegressor
from sklearn.neighbors import RadiusNeighborsRegressor, KNeighborsRegressor
from sklearn.linear_model import LinearRegression, Ridge, BayesianRidge
from sklearn.neural_network import MLPRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
from sklearn.metrics import r2_score
```

```python
def validar(models, X_train, X_test, y_train, y_test):
    for model in models:
        models[model].fit(X_train, y_train)
    predicoes = {}
    for model in models:
        predicoes[model] = round(models[model].score(X_test, y_test), 4)
    return predicoes, models
```

```python
models = {
    'RF': RandomForestRegressor(),
    'GBR':GradientBoostingRegressor(),
    'ETR':ExtraTreesRegressor(),
    'KNR':KNeighborsRegressor(),
    'LR':LinearRegression(),
    'Ridge':Ridge(),
    'BR':BayesianRidge(),
    'DTR':DecisionTreeRegressor(),
    'SVR':SVR(),
    'MLPR':MLPRegressor(),
}
```

```python
predicoes = validar(models, X_train, X_test, y_train, y_test)
```

```python
pd.Series(predicoes[0]).sort_values(ascending=False)
```

```python
models2 = {
    'RF': RandomForestRegressor(),
    'GBR':GradientBoostingRegressor(),
    'ETR':ExtraTreesRegressor(),
    'KNR':KNeighborsRegressor(),
    'LR':LinearRegression(),
    'Ridge':Ridge(),
    'BR':BayesianRidge(),
    'DTR':DecisionTreeRegressor(),
    'SVR':SVR(),
    'MLPR':MLPRegressor(),
}
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=1)
```

```python
X_train_pca = pca.fit_transform(X_train)
```

```python
resul, models2 = validar(models2, X_train_pca, pca.transform(X_test), y_train, y_test)
```

```python
pd.Series(resul).sort_values(ascending=False)
```

```python
models3 = {
    'RF': RandomForestRegressor(),
    'GBR':GradientBoostingRegressor(),
    'ETR':ExtraTreesRegressor(),
    'KNR':KNeighborsRegressor(),
    'LR':LinearRegression(),
    'Ridge':Ridge(),
    'BR':BayesianRidge(),
    'DTR':DecisionTreeRegressor(),
    'SVR':SVR(),
    'MLPR':MLPRegressor(),
}
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train)
resul, models2 = validar(models2, X_train_pca, pca.transform(X_test), y_train, y_test)
```
