```python
pd.DataFrame(dados_faltantes).sort_values('Per. Faltante', ascending=False)
```

## Split

```python
from sklearn.model_selection import train_test_split
```

## Treino

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
def validar(models, X_train, X_test, y_train, y_test):
    for model in models:
        models[model].fit(X_train, y_train)
    predicoes = {}
    for model in models:
        predicoes[model] = round(models[model].score(X_test, y_test), 4)
    return predicoes, models
```

```python
models = {
    'KNC':KNeighborsClassifier(),
    'SVC':SVC(),
    'GPC':GaussianProcessClassifier(),
    'DTC':DecisionTreeClassifier(),
    'RFC':RandomForestClassifier(),
    'MLPC':MLPClassifier(),
    'ABC':AdaBoostClassifier(),
    'GNB':GaussianNB(),
    'QDA':QuadraticDiscriminantAnalysis()
}
```

```python
predicoes = validar(models, X_train, X_test, y_train, y_test)
```

```python
pd.Series(predicoes[0]).sort_values(ascending=False)
```

```python
predicoes[1]['RFC'].score(X_test, y_test)
```
